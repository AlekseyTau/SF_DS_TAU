```python
import pandas as pd
```

```python
ratings = pd.read_csv('/Users/oleksiypolegenko/Projects/IDE/PY_12_Advanced_Methods_In_Pandas/data/ratings_movies.csv')
```

```python
import re 
def get_year_release(arg):
    #находим все слова по шаблону "(DDDD)"
    candidates = re.findall(r'\(\d{4}\)', arg) 
    # проверяем число вхождений
    if len(candidates) > 0:
        #если число вхождений больше 0,
	#очищаем строку от знаков "(" и ")"
        year = candidates[0].replace('(', '')
        year = year.replace(')', '')
        return int(year)
    else:
        #если год не указан, возвращаем None
        return None
```

```python
ratings['year_release'] = ratings['title'].agg(get_year_release).astype('Int64')
```
